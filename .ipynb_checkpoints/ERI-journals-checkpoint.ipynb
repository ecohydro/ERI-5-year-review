```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read in csv of ERI PI publications (09-14 | 14-19) from Dimensions to a single dataframe (all_data)
data_19 = pd.read_csv("data/publications/k-ERI-publications-2014-2019_cleaned.csv",index_col=0,header=0)
#data_14 = pd.read_csv("data/publications/k-ERI-publications-2009-2014_cleaned.csv",index_col=0,header=0)
data_19.head()
```

```python
# Return 15 source titles with most publications (09-14 | 14-19) sorted by size
# Remove all abstracts from 14-19 results
publications_19 = data_19.groupby('Source title').size().to_frame('count').reset_index()
publications_19.drop(publications_19[publications_19['Source title'] == 'Abstracts with programs (Geological Society of America)'].index, inplace = True)
top_pubs_19 = publications_19.nlargest(30, 'count').sort_values(by='count',ascending=False).reset_index()
#publications_14 = data_14.groupby('Source title').size().to_frame('2014').reset_index()
#top_pubs_14 = publications_14.nlargest(15, '2014').sort_values(by='2014',ascending=False)
top_pubs_19
#top_pubs_14
```

```python
# Plot top 15 source titles (09-14 | 14-19) by number of publications in each
top_pubs_19.plot.barh(x='Source title', y='2019',title='ERI Top Journals (2014-2019)',legend=None)
#top_pubs_14.plot.barh(x='Source title', y='2014',title='ERI Top Journals (2009-2014)',legend=None)
plt.savefig('figures/ERI-Py-Top-Journals-14-19-V3.png',bbox_inches='tight',dpi=300)
```

```python
# Return 15 articles with most citations (09-14 | 14-19) sorted by size
#data_19.drop(data_19[data_19['Source title'] == 'M'].index, inplace = True)
top_cited_19 = data_19.nlargest(15, 'Times cited')
#top_pubs_14 = data_14.nlargest(15, '2014').sort_values(by='2014',ascending=False)
top_cited_19
#top_pubs_14
```

```python
# Plot top 15 source titles (09-14 | 14-19) by number of publications in each
top_cited_19.plot.barh(x='Source title', y='Times cited',title='ERI Most Cited Articles (2014-2019)',legend=None)
#top_pubs_14.plot.barh(x='Source title', y='2014',title='ERI Top Journals (2009-2014)',legend=None)
#plt.savefig('figures/ERI-Py-Top-Journals-14-19-V2.png',bbox_inches='tight',dpi=300)
```
